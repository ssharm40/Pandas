# Now we Come to *Pivot Table*

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_excel("salesfunnel.xlsx")
df
```

```python
df.head(4)
```

```python
#This is not strictly required but hels us keep the order we want as we work through the data.

df['Status']= df['Status'].astype('category')
df['Status'].cat.set_categories(['won','pending','presented','declined'],inplace=True)
```

**pivot the data**

We build up the pivot table, I think it's easiest to take it one step at a time.Add items onew at a time and check each step to verify you are getting the result you expect.

The simplest pivot table must have a dataframe and an index. In this case, let's use the name as our index

```python
pd.pivot_table(df,index='Name')

#Normally it's shows only numeric data in front of your index that's why Acc, Pr, Qty is there.
#As some names are already occurs more than one time so it will give u average on their no of arriveness 
#(price sum/no of arrives)
```

```python
# If you want add items on idex (positining will matter)

pd.pivot_table(df,index=['Name','Rep','Manager'])
```

```python
# IF i want to see name throgh Rep
pd.pivot_table(df,index=['Rep','Name','Manager'])
```

```python
#If i want to see through manager name
# These are intersting but not practically useful, What we probably want to do is look at this by Manager and Director.
#It's easy enough to do  by changing the index

pd.pivot_table(df,index=['Manager','Name'])
```

```python
# When others  numerical columns are not useful(quantity ...). Let's remove it by defining the value field
# For sepecific number columns

pd.pivot_table(df,index=['Manager','Name'],values=['Price'])
```

```python
#Fuction on Value Field
#Normally it gives us average but now give the sum

pd.pivot_table(df,index=['Manager','Name'],values='Price', aggfunc=np.sum)
```

```python
#aggfunc can take a list of function. Let's try a mean using the numpy functions and len to get a count

pd.pivot_table(df,index=['Manager','Name'],values='Price',aggfunc=[np.sum,len])
```

```python
pd.pivot_table(df,index=['Manager','Rep'],values=['Price'])
```

```python
# If we want to see sales broken down by tha products, the column variable allow us to define one or more columns

pd.pivot_table(df,index=['Manager','Rep'],values=['Price'],columns=['Product'],aggfunc=[np.sum])
```

```python
# If you want to NaN with 0 value

pd.pivot_table(df,index=['Manager','Rep'],values=['Price'],columns=['Product'],aggfunc=[np.sum],fill_value=0)
```

```python
#I think it should be usefull to add the quantity as well. Add quantity to the value list.
pd.pivot_table(df,index=['Manager','Rep'],values=['Price','Quantity'],columns=['Product'],aggfunc=[np.sum],fill_value=0)
```

You can apply a list of agg function to apply to each value too


```python
table=pd.pivot_table(df,index=['Manager','Status'],columns=['Product'],values=['Quantity','Price'],aggfunc={"Quantity":len,"Price":[np.sum,np.mean]},fill_value=0)
table
```

```python
# Slicing in pivot

table.query('Manager==["Debra Henley"]')
```

# Introduction to Pandas Style API



```python
df=pd.read_excel('2018_Sales_Total_Tabs.xlsx')
df
```

**Group BY**

```python
# A simple group by to look at the performance by customer name

df.groupby('name')['ext price'].agg(['mean','sum'])
```

```python
# To Take decimal 2 point only
# Round Off

df.groupby('name')['ext price'].agg(['mean','sum']).style.format('{0:,.2f}')
```

```python
#To see datewise data on day bases


daily_sales=df.groupby([pd.Grouper(key='date',freq='D')])['ext price'].agg(['sum',len]).reset_index()
daily_sales
```

```python
daily_sales['average']=daily_sales['sum']/daily_sales['len']
daily_sales
```

```python
#To dro a column

#daily_sales.drop('pct of total',axis=1)
```

```python
df['ext price'].sum()
```

```python
#Percentage of total

daily_sales['pct of total']=daily_sales['sum']/df['ext price'].sum()
daily_sales
```

```python
#Now we are giving a setting a format

formate_dic={ 'sum':'${0:,.1f}', 'len':'{:n}'  ,'date': '{:%d-%y}', 'pct_of_total' : '{0:, .2f}' }
```

```python
daily_sales.style.format(formate_dic)
```

```python
daily_sales
```

```python
# Change color of max and min


(daily_sales
.style
 .format(formate_dic)
 .highlight_max(color='lightgreen')
 .highlight_min(color='blue')
)
```

```python
# bar - To color  full
# capion - To give the name of table
#In bar, after color, vmin=0 is used for clearing in color, now do't work in my version

(daily_sales
 .style
 .format(formate_dic)
 .bar(color='pink',subset=['sum'],align='zero')
 .bar(color='lightgreen',subset=['pct of total'],align='zero')
 .set_caption("2018 Sales")

)
```
